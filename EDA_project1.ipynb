```python
def remove_outlier(col):
  sorted(col)
  Q1,Q3=col.quantile([0.25,0.75])
  IQR=Q3-Q1
  lower_range=Q1-(1.5*IQR)
  upper_range=Q3+(1.5*IQR)
  return lower_range,upper_range
```

```python
lrdiscounted_price,urdiscounted_price=remove_outlier(df['discounted_price'])
df['discounted_price']=np.where(df['discounted_price']<lrdiscounted_price,lrdiscounted_price,df['discounted_price'])
df['discounted_price']=np.where(df['discounted_price']>urdiscounted_price,urdiscounted_price,df['discounted_price'])

lractual_price,uractual_price=remove_outlier(df['actual_price'])
df['actual_price']=np.where(df['actual_price']<lractual_price,lractual_price,df['actual_price'])
df['actual_price']=np.where(df['actual_price']>uractual_price,uractual_price,df['actual_price'])

lrdiscount_percentage,urdiscount_percentage=remove_outlier(df['discount_percentage'])
df['discount_percentage']=np.where(df['discount_percentage']<lrdiscount_percentage,lrdiscount_percentage,df['discount_percentage'])
df['discount_percentage']=np.where(df['discount_percentage']>urdiscount_percentage,urdiscount_percentage,df['discount_percentage'])

lrrating,urrating=remove_outlier(df['rating'])
df['rating']=np.where(df['rating']<lrrating,lrrating,df['rating'])
df['rating']=np.where(df['rating']>urrating,urrating,df['rating'])

lrrating_count,urrating_count=remove_outlier(df['rating_count'])
df['rating_count']=np.where(df['rating_count']<lrrating_count,lrrating_count,df['rating_count'])
df['rating_count']=np.where(df['rating_count']>urrating_count,urrating_count,df['rating_count'])

df.shape
```

check for missing value

```python
df.isnull().sum()
```

we replace numercial columns with median values and for categorical column we will replace missing value with mode


```python
df[df.isnull().sum()[df.isnull().sum()>0].index].dtypes
```

replace null values in numerical columns using median

```python
median1=df["discounted_price"].median()
median2=df["actual_price"].median()
median3=df["discount_percentage"].median()
median4=df["rating"].median()
median5=df["rating_count"].median()

df["discounted_price"].replace(np.nan,median1,inplace=True)
df["actual_price"].replace(np.nan,median2,inplace=True)
df["discount_percentage"].replace(np.nan,median3,inplace=True)
df["rating"].replace(np.nan,median4,inplace=True)
df["rating_count"].replace(np.nan,median5,inplace=True)
```

replace null values in categorical columns using mode


```python
 mode1=df["product_name"].mode().values[0]
 mode2=df["category"].mode().values[0]
 mode3=df["about_product"].mode().values[0]

df['product_name']=df['product_name'].replace(np.nan,mode1)
df['category']=df['category'].replace(np.nan,mode2)
df['about_product']=df['about_product'].replace(np.nan,mode3)
```
